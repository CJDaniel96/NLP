```python
import torch
import numpy as np
```

```python
def describe(x):
    print('Type: {}'.format(x.type()))
    print('Shape/size: {}'.format(x.shape))
    print('Values: \n{}'.format(x))
```

# Tensor & Computational Graph

```python
x = torch.ones(2, 2, requires_grad=True)
describe(x)
print(x.grad is None)
```

```python
y = (x + 2) * (x + 5) + 3
describe(y)
print(x.grad is None)
```

```python
z = y.mean()
describe(z)
z.backward()
print(x.grad is None)
```
